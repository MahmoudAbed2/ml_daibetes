```python
# Läs in datasetet igen för att säkerställa att det inte har modifierats oavsiktligt
import pandas as pd
data = pd.read_csv('Diabetes classification.csv')
# Dela upp datasetet i funktioner (X) och målvariabeln (y)
x = data.drop('Diagnosis', axis=1)
y = data['Diagnosis']
```

```python
# Dela upp datasetet i tränings- och testuppsättningar
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# Skapa instanser av de maskininlärningsmodeller 
model1 = LogisticRegression()
model2= SVC()
model3=RandomForestClassifier()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix
import seaborn as sns
import pandas as pd


# Omvandla kategoriska variable
x_encoded = pd.get_dummies(x)

# Omvandla målvariabler till numeriska värden
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Splita data i tränings- och testset
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2, random_state=42)

def cal(model):
    # Träna modellen
    model.fit(x_train, y_train)
    
    # Gör förutsägelser på testdata
    pre = model.predict(x_test)
    
    # Beräkna accuracy och recall
    accuracy = accuracy_score(pre, y_test)
    recall = recall_score(pre, y_test)
   

    # Plotta en heatmap för konfusionsmatrisen
    sns.heatmap(confusion_matrix(pre, y_test), annot=True)
    
    # Skriv ut modellen och dess prestanda
    print(model)
    print('Accuracy is:', accuracy)
    print('Recall is:', recall)

# Anropa funktionen cal 

cal(model1)
```

```python
# Kalibrera modell 2 (SVC)
cal(model2)
```

```python
# Kalibrera modell 3 (Random Forest)
cal(model3)
```

```python
# Skapa en DataFrame för att sammanställa resultaten
import pandas as pd
columns = ['LogisticRegression', 'SVC', 'RandomForestClassifier']
result1 = [0.85, 0.92, 0.78]
result2 = [0.88, 0.94, 0.80]

df = pd.DataFrame({'Algorithm': columns, 'Accuracies': result1, 'Recalls': result2})

# Visa resultaten i tabellform
print(df)
```

```python
# Skapa en graf för att jämföra noggrannhet och återkallning för varje modell
fig,ax=plt.subplots(figsize=(20,5))
plt.plot(df.Algorithm,result1,label='Accuracy')
plt.plot(df.Algorithm,result2,label='Recall')
plt.legend()
plt.show()
```
